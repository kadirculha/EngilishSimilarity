```python
from config import Configurator

cfg = Configurator("comments.json")
```

```python
sentences = list()
for index in range(1,101,1):
    #print(f"{index}. Yorum: \n{cfg.get_yorum(str(index))}")
    sentences.append(cfg.get_yorum(str(index)))
```

```python
import uuid



uuid_list = list()
for index in range(len(sentences)):
    uuid_ = uuid.uuid4()
    print(f"{index}. uuid: {uuid_}")
    uuid_list.append(uuid_)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({"uuid":uuid_list,
                   "Otel Sınıfı":np.array(range(500,600)),
                   "Yorum":sentences})
```

```python
df
```

```python
import json

json_data = {}
for index, row in df.iterrows():
    data = {
        "UUID": str(row["uuid"]),  # Convert UUID to string
        "Otel Sınıfı": int(row["Otel Sınıfı"]),
        "Yorum": row["Yorum"]
    }
    json_data[str(index)] = data

# JSON dosyasına yazma
with open('data.json', 'w', encoding='utf-8') as json_file:
    json.dump(json_data, json_file, ensure_ascii=False, indent=4)

# JSON çıktısını ekrana yazdırma
print(json.dumps(json_data, ensure_ascii=False, indent=4))
```
