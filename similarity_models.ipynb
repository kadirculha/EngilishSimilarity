```python
from config import Configurator
from sentence_transformer import SentenceTransformer
import pandas as pd

cfg = Configurator()
```

```python
pd.set_option('display.max_colwidth', None)
df = pd.DataFrame(cfg.data).T
df
```

```python
type(df.sentences[0])
```

```python
sent1 = []
sent2 = []
for sentences in df.sentences:
    sent1.append(sentences[0])
    sent2.append(sentences[1])
```

```python
df["sent1"] = sent1
df["sent2"] = sent2
df = df[["sent1","sent2"]]
```

# all-MiniLM-L6-v2
2.5m download

```python
l6_v2 = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
l6_v2_list = []
for index, row in df.iterrows():
    score = l6_v2.get_similarity([row["sent1"],row["sent2"]])
    print(f"{index}. Cümle çifti:\n{row}\nSimilarity Score:{score}\n")
    l6_v2_list.append(score)
```

## MPNET 
500mb

```python
mpnet_base_v2 = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')
```
