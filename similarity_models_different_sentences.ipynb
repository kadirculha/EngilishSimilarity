```python
from config import Configurator
from sentence_transformer import SentenceTransformer
import pandas as pd

cfg = Configurator()
```

```python
df = pd.DataFrame(cfg.data).T
df
```

```python
type(df.sentences[0])
```

# all-MiniLM-L6-v2
2.5m download

```python
l6_v2 = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
l6_v2_list = []
for index, sentence_list in df.iterrows():
    score = l6_v2.get_similarity(sentence_list[0])
    print(f"{index}. Cümle çifti:\n{sentence_list[0]}\nSimilarity Score:{score}\n")
    l6_v2_list.append(score)
```

## MPNET
